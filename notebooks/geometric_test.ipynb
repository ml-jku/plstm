```python
import numpy as np
```

```python
NX, NY = 16, 8

a = np.zeros([NX + 1, NY + 1])

for j in range(0, NX + NY - 1):
    for i in range(min(j + 1, NX, NY, NX + NY - j - 1)):
        if j < NX and j < NY:
            x = i + 1
            y = j - i + 1
        elif j >= NX and j >= NY:
            x = j - NY + i + 2
            y = NY - i
        elif j >= NX:
            x = NX - i
            y = j - NX + i + 2
        elif j >= NY:
            x = j - NY + i + 2
            y = NY - i
        print("J", j, "i", i, "x", x, "y", y)
        a[x, y] += 100 * (j + 1) + (i + 1)
```

```python
import matplotlib.pyplot as plt

plt.imshow(a)
```
